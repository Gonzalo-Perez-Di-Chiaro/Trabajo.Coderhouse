```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'https://raw.githubusercontent.com/Gonzalo-Perez-Di-Chiaro/Trabajo.Coderhouse/main/stroke_data%20-%20stroke_data.csv.csv'
data = pd.read_csv(file_path)


sns.set(style="whitegrid")

# Relación entre hipertensión y accidente cerebrovascular
plt.figure(figsize=(10, 6))
sns.countplot(x='hypertension', hue='stroke', data=data, palette='viridis')
plt.title('Relación entre Hipertensión y Accidente Cerebrovascular')
plt.xlabel('Hipertensión')
plt.ylabel('Frecuencia')
plt.xticks(ticks=[0, 1], labels=['No', 'Sí'])
plt.legend(title='Accidente Cerebrovascular', labels=['No', 'Sí'])
plt.show()

# Resumen numérico
hypertension_stroke_summary = data.groupby('hypertension')['stroke'].value_counts(normalize=True).unstack() * 100
print(hypertension_stroke_summary)
```

```python
import pandas as pd


file_path = 'https://raw.githubusercontent.com/Gonzalo-Perez-Di-Chiaro/Trabajo.Coderhouse/main/stroke_data%20-%20stroke_data.csv.csv'
data = pd.read_csv(file_path)

# Identificar valores perdidos
missing_values = data.isnull().sum()

# Mostrar los valores perdidos
print("Valores perdidos por columna:")
print(missing_values)
```

```python
data_cleaned = data.dropna(subset=['sex'])
moda_sex = data['sex'].mode()[0]
data['sex'].fillna(moda_sex, inplace=True)
# Eliminar las filas con valores perdidos en la columna 'sex'
data_cleaned = data.dropna(subset=['sex'])

# Verificar que ya no hay valores perdidos
print("Valores perdidos por columna después de la limpieza:")
print(data_cleaned.isnull().sum())
# Rellenar los valores perdidos en la columna 'sex' con la moda
moda_sex = data['sex'].mode()[0]
data['sex'].fillna(moda_sex, inplace=True)

# Verificar que ya no hay valores perdidos
print("Valores perdidos por columna después de la imputación:")
print(data.isnull().sum())
```
